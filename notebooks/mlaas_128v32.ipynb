# Comparing results

In this notebook we are going to take a look at the results for N=128 and N=32 on the MLaaS model.

```python
import os
os.chdir('../')
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
csv_128 = 'results/mlaascorr_128.csv'
csv_32 = 'results/mlaascorr_32.csv'
```

```python
df_128 = pd.read_csv(csv_128)
display(df_128)

df_32 = pd.read_csv(csv_32)
display(df_32)
```

```python
sns.set_style('whitegrid')
sns.set(font_scale=1.1)
plt.figure(figsize=(8, 8), dpi=300)

# Show regression plot
# plt.ylim(0, 1)

# metric = 'L-inf'
metric = 'Cosine'

# s = sns.regplot(data=df_128, y='Transfer', x='Linf', truncate=False, label='N = 128')
s = sns.regplot(data=df_128, y='Transfer', x='Cosine', truncate=False, label='N = 128')
# sns.regplot(data=df_32, y='Transfer', x='Linf', truncate=False, label='N = 32', ax=s)
sns.regplot(data=df_32, y='Transfer', x='Cosine', truncate=False, label='N = 32', ax=s)

s.set(title='Transfer vs. Zest {} distance'.format(metric))
plt.legend()
plt.show()
```
